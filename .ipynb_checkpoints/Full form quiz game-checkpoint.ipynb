```python
print("Welcome to my computer quize.")

playing = input("Do you want to play? ")

if playing.lower() != 'yes':
    exit()
    
else:
    print("Okay! Let's play : ")
    score = 0
    
    answer = input("What is the full form of CPU: ")
    if answer.lower() == "central processing unit":
        score += 1
    
    else:
        print("Incorrect")
        
    answer = input("What does SMPT stand for: ")
    if answer.lower() == "switch mode power supply":
        score += 1
    
    else:
        print("Incorrect")
        
    answer = input("What does FTP stand for: ")
    if answer.lower() == "file transfer protocol":
        score += 1
        
    else:
        print("Incorrect")
        
    answer = input("What does BIOS stand for: ")
    if answer.lower() == "basic input output system":
        score += 1
        
    else:
        print("Incorrect")
    
    answer = input("What does CD stand for: ")
    if answer.lower() == "compact disk":
        score += 1
        
    else:
        print("Incorrect")
    
    answer = input("What does DVD stand for: ")
    if answer.lower() == "digital video disk":
        score += 1
        
    else:
        print("Incorrect")
    
    answer = input("What does HDD stand for: ")
    if answer.lower() == "hard disk drive":
        score += 1
        
    else:
        print("Incorrect")
        
    answer = input("What does HDMI stand for: ")
    if answer.lower() == "high definition multimedia interface":
        score += 1
        
    else:
        print("Incorrect")
    
    answer = input("What does LCD stand for: ")
    if answer.lower() == "liquid crystal display":
        score += 1
        
    else:
        print("Incorrect")
    
    answer = input("What does LED stand for: ")
    if answer.lower() == "light emitting diode":
        score += 1
        
    else:
        print("Incorrect")

print(f"Your score is {score}.")
```
